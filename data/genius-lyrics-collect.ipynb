```python
import lyricsgenius as lg
import os
from dotenv import load_dotenv
import pandas as pd
import re

load_dotenv()

GENIUS_CID = os.environ.get("GENIUS_CID")
GENIUS_SECRET = os.environ.get("GENIUS_SECRET")
GENIUS_ACCESS_TOKEN=os.environ.get("GENIUS_ACCESS_TOKEN")
```

```python
discography = pd.read_csv("beatles_set.csv")
artist = "The Beatles"
```

```python
discography
```

```python
genius = lg.Genius(GENIUS_ACCESS_TOKEN, skip_non_songs=True, 
                   excluded_terms=["(Remix)", "(Live)"], remove_section_headers=True)
```

```python
lyrics_list = []

for track in discography['track']:
    try:
        lyrics = genius.search_song(track, artist, get_full_info=True).lyrics
        lyrics_list.append(lyrics.replace('EmbedShare URLCopyEmbedCopy', ''))
    except:
        lyrics_list.append('not found')
    
discography['lyrics'] = lyrics_list
```

```python
discography.to_csv("beatles_full.csv", index=False)
```

```python
discography
```

```python
def clear_lyrics(lyrics):
    lyrics = re.sub('\n', '. ', lyrics)
    lyrics = re.sub(r'\([^)]*\)', '', lyrics)
    return lyrics
```

```python
discography['lyrics'] = discography['lyrics'].apply(lambda x: clear_lyrics(x))
```
